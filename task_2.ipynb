```python
import pandas as pd
```

#### 1. Чтение dataframe из файла

```python
df = pd.read_csv("horror_movies.csv")
df
```

#### 2. Проверка на наличие пустых значений

```python
df.isna().sum()
```

Удаление строк, содержащих пустые значения

```python
df = df.dropna()
df
```

#### 3. Фильтрация данные, чтобы в выборке остались только четные id и статус фильма был "Released".

```python
df_filtered = df.loc[(df['id'] % 2 == 0) & (df['status'] == 'Released')]
df_filtered
```

#### 4. Вывести сумму доходов (revenue) для каждого жанра (genre_names).

```python
df_filtered.loc[:, 'genre_names'] = df_filtered['genre_names'].str.split(',').apply(lambda x: [genre.strip() for genre in x])
df_exploded = df_filtered.explode('genre_names')
revenue_by_genre = df_exploded.groupby('genre_names')['revenue'].sum()
revenue_by_genre
```

#### 5. Для каждого жанра посчитать отношение оценки (vote_average) и бюджета (budget). Использовать средние значения.

```python
avg_vote_budget_ratio = (df_exploded.groupby('genre_names')['vote_average'].mean() / 
                                 df_exploded.groupby('genre_names')['budget'].mean())
avg_vote_budget_ratio
```
